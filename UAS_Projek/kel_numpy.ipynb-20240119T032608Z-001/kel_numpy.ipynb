D. PEMODELAN

```python
#Import sklearn untuk modeling data

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

from sklearn.naive_bayes import GaussianNB
```

**Naive** **Bayes**

```python
#Pengecekan apakah typedata sudah menjadi numeric

dataset.head()
```

```python
x = dataset.iloc[:, :8].values
y = dataset.iloc[:, 2].values
```

```python
print (x)
```

```python
print (y)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=123)

print("x_train = ", len(x_train))
print("x_test = ", len(x_test))
print("y_train = ", len(y_train))
print("y_test = ", len(y_test))
```

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

KLASIFIKASI

```python
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

PREDIKSI

```python
y_prediksi = classifier.predict(x_test)
y_prediksi
```

PROBABILITAS

```python
classifier.predict_proba(x_test)
```

```python
cm = confusion_matrix(y_test, y_prediksi)
print(cm)
```

AKURASSI

```python
akurasi = classification_report(y_test,y_prediksi)
print(akurasi)
```

```python
akurasi = accuracy_score(y_test,y_prediksi)
print("Tingkat Akurasi : %d persen"%(akurasi*100))
```
